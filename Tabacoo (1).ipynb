 # ALL Diagnosis Types in ANOVA Testing  	(Hypothesis Testing)

 

```python
from scipy.stats import f_oneway

# Create a list of Value arrays for each Diagnosis Type
groups = [df_stat[df_stat['Diagnosis Type'] == diag]['Value'] 
          for diag in df_stat['Diagnosis Type'].unique()]

# Run one-way ANOVA across all diagnosis types
f_stat, p_val = f_oneway(*groups)

print(f"F-statistic: {f_stat:.2f}, P-value: {p_val:.4f}")
```

# Interpertation: Since different diagnosis types show very different averages, we should not treat them the same. 
# Recommendation :Resources and policies must be tailored to each disease category, focusing more on those with higher fatality or admission rate.


# *Ttest, p-value*


```python
from scipy.stats import ttest_ind

# Select rows where Diagnosis Type contains 'smoking'
smoking = df_stat[df_stat['Diagnosis Type'].str.contains('smoking')]['Value']

# Select rows where Diagnosis Type does NOT contain 'smoking'
non_smoking = df_stat[~df_stat['Diagnosis Type'].str.contains('smoking')]['Value']

# Run independent t-test on both groups
t_stat, p_val = ttest_ind(smoking, non_smoking, equal_var=False)

# Print test results
print(f"T-statistic: {t_stat:.2f}, P-value: {p_val:.4f}")
```

  # Interpertaion: Smoking patients show significantly lower average admissions compared to nonsmoking patients, and this difference is statistically real. 
  # Action: healthcare programs should prioritize awareness and prevention for smoking groups, while allocating more resources to nonsmoking categories where admissions are higher.

 # Independent Samples TTest (Before vs After 2012)

```python
df = df_pres.copy()
```

```python
from scipy import stats

# Split groups manually based on fiscal year
before = df[df["Year"].isin(["2010/11","2011/12"])]["Cost Per Prescription"]
after = df[df["Year"].isin(["2012/13","2013/14","2014/15"])]["Cost Per Prescription"]

print("Mean Cost Before 2012:", before.mean())
print("Mean Cost After 2012:", after.mean())

t_stat, p_val = stats.ttest_ind(before, after, equal_var=False)
print("T-statistic:", t_stat)
print("P-value:", p_val)
```

# interpertation:Prescription costs went up after 2012  from about 26 to 27 rupees each. The increase is visible, but statistically its not strong enough to confirm with high confidence because the sample size is small.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Step 1: Melt the data (wide  long format)
df_melted = df.melt(
    id_vars=["Year"], 
    value_vars=["16 and Over", "16-24", "25-34", "35-49", "50-59", "60 and Over"],
    var_name="Age Group", 
    value_name="Smoking Rate"
)

# Step 2: Line chart with Seaborn
plt.figure(figsize=(10,6))
sns.lineplot(x="Year", y="Smoking Rate", hue="Age Group", data=df_melted, marker="o")

plt.title("Smoking Rate Trend by Age Group")
plt.xlabel("Year")
plt.ylabel("Smoking Rate (%)")
plt.legend(title="Age Group")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Interpretation:
# Smoking rates have steadily declined across all age groups over time. Adults aged 2534 and 3549 had the highest rates in earlier decades, while older adults (60+) consistently showed the lowest.
 # Recomnndation:
 # Focus cessation programs on adults aged 2549, while reinforcing prevention among younger groups.

```python

df_melted = df.melt(
    id_vars=["Year","Sex","Method"], 
    value_vars=["16 and Over","16-24","25-34","35-49","50-59","60 and Over"],
    var_name="Age Group",
    value_name="Smoking Rate"
)
# Step 3: Aggregate average smoking rate by Sex
sex_avg = df_melted.groupby("Sex")["Smoking Rate"].mean()

# Step 4: Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(
    sex_avg,
    labels=sex_avg.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=["skyblue", "lightcoral", "lightgreen"]
)
plt.title("Smoking Rate Distribution by Sex")
plt.show()
```

# Interpretation:
# The dataset shows a nearly even split between male and female categories, but a large proportion (33.2%) is recorded as Unknown, indicating incomplete demographic information.
# Recommendation:
# Future surveys should reduce the share of Unknown entries by improving data collection methods, ensuring clearer gender classification. This will strengthen demographic accuracy and allow more reliable, targeted public health analysis.
